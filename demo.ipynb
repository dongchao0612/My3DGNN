```python
output=torch.randn([32, 14, 640, 480])
output
```

```python
pred = output.permute(0, 2, 3, 1).contiguous()
print(pred.shape)
print("len(pred)",len(pred))
pred = pred.view(-1, 14)
pred
```

```python
pred_max_val, pred_arg_max = pred.max(1)

```

```python
pred_max_val

```

```python
pred_arg_max

```

```python
target=torch.randn([2,640,480])
target
```

```python
pairs = target.view(-1) * 14 + pred_arg_max.view(-1)  #[ 0]
pairs
```

```python
confusion_matrix = torch.cuda.FloatTensor(np.zeros(14 ** 2))
# 计算混淆矩阵
for i in range(14 ** 2):
    cumu = pairs.eq(i).float().sum()
    confusion_matrix[i] += cumu.item()
```

```python
confusion_matrix
```

```python
confusion_matrix = confusion_matrix.reshape((14, 14))
class_iou = np.zeros(14)
confusion_matrix[0, :] = np.zeros(14)
confusion_matrix[:, 0] = np.zeros(14)

# 计算交并比
for i in range(1, 14):
    class_iou[i] = confusion_matrix[i, i] / (np.sum(confusion_matrix[i, :]) + np.sum(confusion_matrix[:, i]) - confusion_matrix[i, i])
```
